```python
#Dataframe of all 8th grade scores in 2009

df_grade8 = df.loc[(df['Grade'] == '8') & (df["Year"].isin(["2009"]))]
df_grade8
```

```python
#Bar graph of percent of each student group at meeting learning standards(level 3+4)
fig = px.bar(df_grade8, x="Level 3+4 %", y="Ethnicity", orientation='h')
fig.show()
```

```python
#Percentage of each ethnicity that took the state math test in the 8th grade in 2009

all_students = df_grade8['Number Tested'].sum()
meeting_standard = df_grade8['Level 3+4 #'].sum()

student_percent = (meeting_standard/all_students)

print(student_percent, 'Percent of 8th graders met math learning standard in 2009')
```

```python
#pie chart of percentage enthicities that took the 8th grade state math test in 2009
fig = px.pie(df_grade8, values='Number Tested', names='Ethnicity', color_discrete_sequence=px.colors.sequential.RdBu_r)
fig.show()
```

```python
#Dataframe of all 5th graders in 2010
df_grade5 = df.loc[(df['Grade'] == '5') & (df["Year"].isin(["2010"]))]
df_grade5.head()
```

```python
#Percent of students that met math learning standard
all_grade5 = df_grade5['Number Tested'].sum()
passing_grade5 = df_grade5['Level 3+4 #'].sum()
grade5_percent = (passing_grade5/all_grade5)

print(grade5_percent, "Percent of 5th graders met math learning standard in 2010")
```

It is evident in the line graphs and percentages that across all ethnicities, the math test scores plummetted in 2010 with very little recovery in 2011.
